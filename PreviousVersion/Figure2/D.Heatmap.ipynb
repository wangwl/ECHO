```python
import os
import re
import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dmr_mCG_df = pd.read_csv("consensus_DMRs.mCG.tsv", sep="\t")
```

```python
cell_order = ['B-Mem','B-Naive', 'Th-Mem','Th-Naive','Tc-Naive','NK-cell1', 'NK-cell2', 'Tc-Mem', 'Monocyte']
```

```python
dmr_mCG_df.columns = [re.sub("methylation_level_", "", x) for x in dmr_mCG_df.columns]
```

```python
fig, ax = plt.subplots(1,1, figsize=(6, 10))
sns.heatmap(dmr_mCG_df[cell_order], cmap='viridis', ax=ax, yticklabels=False, linewidths=0, rasterized=True)
plt.savefig("DMR_mCG.pdf")
```
