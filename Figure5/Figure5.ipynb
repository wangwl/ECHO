## Figure 5A

```python
gDMRs = pd.read_csv("all_gDMRs.bed", sep="\t", names=['chrom', 'start','end','cell', 'gDMR_Type'])
```

```python
count_stat = gDMRs[['gDMR_Type','cell']].value_counts().to_frame().reset_index().rename({0: 'count'}, axis=1)
cis = count_stat[count_stat['gDMR_Type']=='cis']
```

```python
fig, axes = plt.subplots(1,1,figsize=(4, 3))

ax = sns.barplot(data=cis.sort_values(by='cell'), x='cell', y='count', hue='cell', palette=cell2color)
ax.xaxis.set_visible(False)
ax.set_ylabel(f'gDMR Counts')
plt.savefig('Figure5A.gDMR_count.pdf')
```

## Figure 5B

```python
filtered_result = pd.read_csv("gDMR_over_eDMR.enrich.csv")
```

```python
df_col = filtered_result[['DMR']].set_index('DMR')
df_col['Celltype'] = [x.split("_")[1] for x in df_col.index]


df_row = filtered_result[['feature']].set_index('feature').drop_duplicates()
df_row[['Feature', 'Celltype']] = [x.split("_") for x in df_row.index]
# df_row['Feature'] = ['Genomic_' + x if not x.startswith('H3') else x for x in df_row['Feature']]
df_row = df_row.drop_duplicates()
df_col = df_col.drop_duplicates()
```

```python
y_order = sorted(df_row.index)
```
